```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Titanic_LR_train.csv",delimiter=",")
data
```

```python
del data["Name"]
del data["Ticket"]
del data["Cabin"]
```

```python
#data.Cabin.unique()
```

```python
def gender(s):
    if s == "male":
        return 1
    else:
        return 2
data["Gender"] = data.Sex.apply(gender)
data
```

```python
def emb(s):
    if s=='S':
        return 1
    elif s=='Q':
        return 2
    else:
        return 3
data["Emb"] = data.Embarked.apply(emb)
data
```

```python
del data["Sex"]
del data["Embarked"]
```

```python
data.fillna(value = data["Age"].mean(),inplace=True)
data
```

```python
data["SurvivedF"] = data["Survived"]
data
```

```python
del data["Survived"]
data
```

```python
data_np = data.to_numpy()
X=data_np[:,:-1]
Y=data_np[:,-1]

from sklearn.linear_model import LogisticRegression
alg = LogisticRegression()
alg.fit(X,Y)
```

```python
test_data = pd.read_csv("Titanic_LR_test.csv",delimiter=",")
test_data
```

```python
del test_data["Name"]
del test_data["Ticket"]
del test_data["Cabin"]
test_data
```

```python
test_data.fillna(value = test_data["Age"].mean(),inplace=True)
test_data
```

```python
def gender(s):
    if s == "male":
        return 1
    else:
        return 2
test_data["Gender"] = test_data.Sex.apply(gender)
del test_data["Sex"]
test_data
```

```python
def emb(s):
    if s=='S':
        return 1
    elif s=='Q':
        return 2
    else:
        return 3
test_data["Emb"] = test_data.Embarked.apply(emb)
del test_data["Embarked"]
test_data
```

```python
alg.score(X,Y)
```

```python
y_pred = alg.predict(test_data.to_numpy())
y_pred
```

```python
alg.score(test_data.to_numpy(),y_pred)
```

```python
np.savetxt("Titanic_LR_pred.csv",y_pred,delimiter=",")
```
